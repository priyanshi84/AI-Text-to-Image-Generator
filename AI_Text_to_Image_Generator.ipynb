### ğŸ› ï¸ Step 4: Load the Stable Diffusion Model
Load the `StableDiffusionPipeline` using the `runwayml/stable-diffusion-v1-5` model.

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    use_safetensors=True
)
pipe = pipe.to("cuda")
```

### ğŸ¨ Step 5: Prompt for Image Generation
Take text input from the user that describes the image to generate.

```python
user_prompt=input("Enter your image prompt:")
```

### ğŸ–¼ï¸ Step 6: Generate and Display the Image
Use the pipeline to generate an image from the user prompt.
Then display the generated image using matplotlib.

```python
image = pipe(user_prompt).images[0]
#Display the generated image
plt.imshow(image)
plt.axis("off")
plt.title("Generated Image")
plt.show()
```
